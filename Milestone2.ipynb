```python
# Your Solution here
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/Users/takonyabadza/Desktop/course-project-group_317/spiglobalrankings.csv")
df['weighted_scores']=df['off']*.25 + df['def']*.25 + df['spi']*.50
df2 = df.groupby(by=['spi']).mean().sort_values(by=['def'], ascending=False)
sns.set_theme(style="dark", font_scale=1.5)
d = sns.boxplot(x='def', y='name', data=df)
d = sns.stripplot(x='def', y='name', data=df)
plt.title('Teams sorted by Defense')
d.set(xlabel='Defense', ylabel=None)
plt.show()
```
